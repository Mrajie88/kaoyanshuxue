# 爬取京东数据

```python
# 张宇
pages = 100
for i in range(pages):
    url1 = "https://sclub.jd.com/comment/productPageComments.action?callback=fetchJSON_comment98vv648&productId=12507770&score=3&sortType=5&page="
    url2 = str(i)
    url3 = "&pageSize=10&isShadowSku=0&fold=1"
    url = url1+url2+url3
    #构建请求头
    headers = {"user-Agent":"Mozilla/5.0","Referer":"https://item.jd.com/12507770.html"}
    try:
        r = requests.get(url,headers = headers)
    except:
        print("第"+url2+"页爬取失败")
    #获取和处理json数据
    text = r.text[25:-2]
    comment_json=json.loads(text)
    #初始化数据框
    comment_data = pd.DataFrame(columns=["id","content","score","time"])
    for j in range(len(comment_json["comments"])):
        if(comment_json["comments"][j]["id"] !=None):
            Userid  = comment_json["comments"][j]["id"]
        else:
            Userid  = ""
        if(comment_json["comments"][j]["content"]!=None):
            content = comment_json["comments"][j]["content"]
        else:
            content = ""
        if(comment_json["comments"][j]["score"]!=None):
            score = comment_json["comments"][j]["score"]
        else:
            score = ""
        if(comment_json["comments"][j]["creationTime"]!=None):
            Comment_time = comment_json["comments"][j]["creationTime"]
        else:
            Comment_time = ""
        item = {"id":Userid,"content":content,"score":score,"time":Comment_time}
        comment_data = comment_data.append(item,ignore_index=True)
    comment_data.to_csv("comment_zhangyu123.csv",encoding='utf_8_sig',mode="a+",header=False)
    #控制爬取速度
    time.sleep(5)
```

```python
#李永乐
# 制作语料 好评、中评、差评混合
liyongle_1 = pd.read_csv("comment_liyongle_1.csv",header=None)
liyongle_1.columns = ["Userid","content","score","time"]
print(len(liyongle_1))

liyongle_2 = pd.read_csv("comment_liyongle_2.csv",header=None)
liyongle_2.columns = ["Userid","content","score","time"]
print(len(liyongle_2))

liyongle_3 = pd.read_csv("comment_liyongle_3.csv",header=None)
liyongle_3.columns = ["Userid","content","score","time"]
print(len(liyongle_3))
```

```python
# 混合比例 99% 0.5% 0.5% 495 3 3 实验结果取5次平均成绩
sentiment_list = []
for i in range(5):
    temp1 = liyongle_1.sample(n=3)
    temp2 = liyongle_2.sample(n=3)
    temp3 = liyongle_3.sample(n=495)
    temp = [temp1,temp2,temp3]
    res = pd.concat(temp)
    for j in range(len(res)):
        text = res.iloc[j,1]
        s = SnowNLP(text)
        sentiment_list.append(s.sentiments)
res = np.array(sentiment_list).mean()
res
```
